```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


```

to convert dates directly at import add:
- parse_dates=['col_name']

can also make the date the index of the df
- parse_dates=['col_name'], index_col='col_name'

```python
day_import = pd.read_csv('day.csv', parse_dates=['dteday'])
hour_import = pd.read_csv('hour.csv', parse_dates=['dteday'])
```

```python
day=day_import.copy()


day.loc[day['holiday'] == 0, 'is_holiday'] = 'No'
day.loc[day['holiday'] == 1, 'is_holiday'] = 'Yes'




day.loc[day['workingday'] == 0, 'is_workday'] = 'No'
day.loc[day['workingday'] == 1, 'is_workday'] = 'Yes'

day.loc[day['season'] == 1, 'season'] = 'Winter'
day.loc[day['season'] == 2, 'season'] = 'Spring'
day.loc[day['season'] == 3, 'season'] = 'Summer'
day.loc[day['season'] == 4, 'season'] = 'Fall'

#day['dteday']= pd.to_datetime(day['dteday'])


day['atemp']=day['atemp']*(50+16)-16 #(t-t_min)/(t_max-t_min), t_min=-16, t_max=+50

# we are creating a new column called "temp_label", we define a range and associate that range with a label
day['temp_label'] = day['atemp'].apply(lambda value: 'cold'
if value <= 10 else 'warm'
if value <= 30 else 'hot')




print(day.describe())


```

## Visualise

```python
sns.lineplot(x='mnth',y='cnt' , data=day, hue='yr' )
plt.title('Count per month',  fontsize='20')
plt.ylabel("Count")
plt.xlabel("Month")
```

```python
def plot_df(df, x, y, title="", xlabel='Date', ylabel='Count', dpi=100):
    plt.figure(figsize=(16,5), dpi=dpi)
    plt.plot(x, y, color='tab:red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()

plot_df(day, x=day.dteday, y=day.cnt, title='Bike count per day')    
```

## Seasonal plot

```python


# Draw Plot
fig, axes = plt.subplots(1, 2, figsize=(20,7), dpi= 80)
sns.boxplot(x='yr', y='cnt', data=day, ax=axes[0], hue='yr', palette='hsv', legend=False)
sns.boxplot(x='mnth', y='cnt', data=day, hue='mnth', palette='hsv', legend=False)

# Set Title
axes[0].set_title('Year-wise Box Plot\n(The Trend)', fontsize=18); 
axes[1].set_title('Month-wise Box Plot\n(The Seasonality)', fontsize=18)
plt.show()
```

```python
day_heatmap=day_import.copy()



if 'dteday' in day_heatmap:
    day_heatmap=day_heatmap.drop(labels='dteday', axis=1)

mask = np.triu(np.ones_like(day_heatmap.corr()))

ax=sns.heatmap(
    day_heatmap.corr(), 
    annot=True,fmt=".1f",
    square=True,
    mask=mask,
     cmap='vlag',
    vmin = -1,
    vmax = 1
)


ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
)
sns.set(rc = {'figure.figsize':(15,10)})
```
